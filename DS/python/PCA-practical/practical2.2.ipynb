### q2

```python
from sklearn.decomposition import PCA
pca = PCA()
df_1=pd.DataFrame(data=pca.fit_transform(Dataset_3))
df_1
```

```python
df_1.index = Dataset_3.index
```

```python
df_1
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum(),marker='o', linestyle='-')
plt.axhline(y=0.9, color='b', linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel("The Number of Components")
plt.ylabel("Cumulative Explained Variances" )
plt.show()
```

```python
df_pca_components = pca.explained_variance_ratio_.cumsum() > 0.90
df_pca_components
```

```python
df_pca_components.sum() #13個是達到的0.9之後加上的 所以false的是加起來等於0.9
```

```python
Number_of_components = 17 - df_pca_components.sum()
Number_of_components
```

create new dataset for new features

```python
pca = PCA(n_components=Number_of_components)

# clean up column names
# give column names to dataframe
columns = ['PCA %i' % i for i in range(Number_of_components)]
score_pca = pca.fit_transform(Dataset_3)
# create PCA dataframe
X_scaled_Display = pd.DataFrame(pca.fit_transform(Dataset_3),columns=columns)
X_scaled_Display
```
